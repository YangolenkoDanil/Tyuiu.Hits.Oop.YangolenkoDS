Создать базовый класс Ticket в C#, который будет представлять билеты на
различные мероприятия. На основе этого класса разработать 2-3 производных
класса, демонстрирующих принципы наследования и полиморфизма. В каждом из
классов должны быть реализованы новые атрибуты и методы, а также
переопределены некоторые методы базового класса для демонстрации
полиморфизма.


#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

<h2 style="color:DodgerBlue">Реализация:</h2>

```
public interface ITicket
{
    string EventLocation {get; set;}
}

public class Ticket : ITicket
{
    public int TicketId { get; set; }
    public int RowNumber { get; set; }
    public int SeatNumber { get; set; }
    public string EventDate {get; set;}
    public string AgeLimit {get; set;}
    public string ContactNumber {get; set;}
    public string EventLocation {get; set;}

    public virtual void PrintTicketInfo()
    {
        Console.WriteLine($"Билет {TicketId}: ряд {RowNumber}, место {SeatNumber}, Дата мероприятия: {EventDate}, Место проведения: {EventLocation}, Возрастное ограничение: {AgeLimit}, Контактный телефон: {ContactNumber}");
    }

    public virtual void ReserveSeat()
    {
        Console.WriteLine("Место забронировано.");
    }

    public virtual string GetTicketDetails()
    {
        return $"ID: {TicketId}, Ряд: {RowNumber}, Место: {SeatNumber}, Дата мероприятия: {EventDate}, Место проведения: {EventLocation}, Возрастное ограничение: {AgeLimit}, Контактный телефон: {ContactNumber}";
    }

    public virtual void BuyingTicket()
    {
        Console.WriteLine("Билет куплен");
    }

    public virtual void TicketRefund()
    {
        Console.WriteLine("Билет возвращен");
    }

    public Ticket(int ticketId, int rowNumber, int seatNumber, string eventDate, string ageLimit, string contactNumber, string eventLocation)
    {
        TicketId = ticketId;
        RowNumber = rowNumber;
        SeatNumber = seatNumber; 
        EventDate = eventDate;
        AgeLimit = ageLimit;
        ContactNumber = contactNumber;
        EventLocation = eventLocation;
    }
}

public class ConcertTicket : Ticket, ITicket
{
    public string ArtistName { get; set; }
    public string TourName {get; set;}
    public string EventLocation {get; set;}   

    public override void PrintTicketInfo()
    {
        base.PrintTicketInfo();
        Console.WriteLine($"Исполнитель: {ArtistName}");
    }

    public override void BuyingTicket()
    {
        Console.WriteLine($"Билет на концерт {ArtistName} куплен");
    }

    public override void TicketRefund()
    {
        Console.WriteLine($"Билет на концерт {ArtistName} возвращен");
    }

    public override void ReserveSeat()
    {
        base.ReserveSeat();
    }

    public ConcertTicket(int ticketId, int rowNumber, int seatNumber, string eventDate, string ageLimit, string contactNumber, string artistName, string tourName, string eventLocation)
        : base(ticketId, rowNumber, seatNumber, eventDate, ageLimit, contactNumber, eventLocation)
    {
        ArtistName = artistName;
        TourName = tourName;
        EventLocation = eventLocation;        
    }
}

public class PlayTicket : Ticket, ITicket
{
    public string PlayTitle { get; set; }
    public string EventLocation {get; set;}
    public string MainCharacters {get; set;}
    public string Actors {get; set;}

    public override string GetTicketDetails()
    {
        var details = base.GetTicketDetails();
        return $"{details} Спектакль: {PlayTitle}";
    }
    
    public override void BuyingTicket()
    {
        Console.WriteLine($"Билет на спектакль {PlayTitle} куплен");
    }

    public override void TicketRefund()
    {
        Console.WriteLine($"Билет на спектакль {PlayTitle} возвращен");
    }

    public override void ReserveSeat()
    {
        base.ReserveSeat();
    }

    public PlayTicket(int ticketId, int rowNumber, int seatNumber, string eventDate, string ageLimit, string contactNumber, string playTitle, string eventLocaion, string mainCharacters, string actors)
        : base(ticketId, rowNumber, seatNumber, eventDate, ageLimit, contactNumber, eventLocaion)
    {
        PlayTitle = playTitle;
        EventLocation = eventLocaion;
        MainCharacters = mainCharacters;
        Actors = actors;
    }
}

public class ConferenceTicket : Ticket, ITicket
{
    public string ConferenceName { get; set; }
    public string EventLocation {get; set;}
    public string ConferenceHost {get; set;}
    public string Speaker {get; set;}
    
    public override void ReserveSeat()
    {
        base.ReserveSeat();
    }

    public override void BuyingTicket()
    {
        Console.WriteLine($"Билет на {ConferenceName} куплен");
    }

    public override void TicketRefund()
    {
        Console.WriteLine($"Билет на {ConferenceName} возвращен");
    }

    public ConferenceTicket(int ticketId, int rowNumber, int seatNumber, string eventDate, string ageLimit, string contactNumber, string conferenceName, string eventLocation, string conferenceHost, string speaker)
        : base(ticketId, rowNumber, seatNumber, eventDate, ageLimit, contactNumber, eventLocation)
    {
        ConferenceName = conferenceName;
        EventLocation = eventLocation;
        ConferenceHost = conferenceHost;
        Speaker = speaker;
    }
}

public class ClosedConference : ConferenceTicket
{
    public string Members {get; set;}

    public ClosedConference(int ticketId, int rowNumber, int seatNumber, string eventDate, string ageLimit, string contactNumber, string conferenceName, string eventLocation, string conferenceHost, string speaker, string members)
        : base(ticketId, rowNumber, seatNumber, eventDate, ageLimit, contactNumber, conferenceName, eventLocation, conferenceHost, speaker)
    {
        Members = members;
    }
}

ConcertTicket myConcertTicket = new ConcertTicket(1, 2, 16, "19.10.2024", "16+", "89350002535", "OG Buda", "DrillTour", "Дом культуры");
PlayTicket myPlayTicket = new PlayTicket(2, 5, 32, "21.01.2025", "12+", "88005553535", "Ромео и Джульетта", "ДрамТеатр", "Ромео, Джульетта", "Иванов Иван, Олеся Евгеньевна");
ConferenceTicket myConferenceTicket = new ConferenceTicket(3, 7, 52, "30.09.2024", "16+", "87829941234", "IT конференция", "Проспект Мира, 80", "Жучков Владимир", "Стас Михайлов");
ClosedConference myClosedConference = new ClosedConference(4, 4, 42, "20.11.2024", "18+", "87992023445", "Обсуждение планов компании", "Главный офис, кабинет 35", "Дмитрий Иванов", "Дмитрий Иванов", "Данила Петров, Евгений Простой, Никита Собчак");

myConcertTicket.PrintTicketInfo();
myConcertTicket.BuyingTicket();
myConcertTicket.ReserveSeat();
myConcertTicket.TicketRefund();

myPlayTicket.PrintTicketInfo();
myPlayTicket.BuyingTicket();
myPlayTicket.ReserveSeat();
myPlayTicket.TicketRefund();

myConferenceTicket.PrintTicketInfo();
myConferenceTicket.BuyingTicket();
myConferenceTicket.ReserveSeat();
myConferenceTicket.TicketRefund();

myClosedConference.PrintTicketInfo();
myClosedConference.BuyingTicket();
myClosedConference.ReserveSeat();
myClosedConference.TicketRefund();
```
